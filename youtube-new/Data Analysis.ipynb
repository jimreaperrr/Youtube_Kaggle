```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud
```

```python
df = pd.read_csv("USvideos.csv")
df
```

```python
df2 = df.sort_values(by=["views"], ascending= False)

df2
```

```python
top_iphone = df2[df2['title'].str.contains('iPhone' or 'Apple')].sort_values(by='views', ascending=False)
top = top_iphone.groupby('channel_title')['views','tags'].max()
top = top.reset_index().sort_values("views",ascending=False)
#top.sort_values("views",ascending=False).head(10)
```

```python
top
```

```python
ax = top.plot(kind='bar',x='channel_title',y='views',title='Top10 iPhone channels')
ax.set_xlabel('Channel Title')
ax.set_ylabel('Views(in m)')
plt.show()
```
